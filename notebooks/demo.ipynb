# EmoNarrify - End-to-End Demo
Run the full pipeline: Image -> Emotional Story Audio

```python
# ===== Setup =====
# !git clone https://github.com/YOUR_REPO/emonarrify.git
# %cd emonarrify
# !pip install -r requirements.txt

import sys, os
sys.path.insert(0, os.path.abspath('..'))

from PIL import Image
from emonarrify import EmoNarrifyPipeline
```

```python
# ===== Initialize Pipeline =====
# Stub mode (all dummy):
pipeline = EmoNarrifyPipeline()

# Real mode (after all phases are trained):
# pipeline = EmoNarrifyPipeline(
#     phase1_adapter='weights/phase1_lora/',
#     phase2_mlp='weights/phase2_mlp.pt',
#     phase3_tts='weights/phase3_tts.pt',
# )

# Fallback mode (skip Phase 2, use discrete label):
# pipeline = EmoNarrifyPipeline(
#     phase1_adapter='weights/phase1_lora/',
#     phase3_tts='weights/phase3_tts.pt',
#     use_fallback=True,
# )
```

```python
# ===== Run Inference =====
image = Image.new('RGB', (224, 224), color='gray')  # replace with real image
# image = Image.open('path/to/your/image.jpg').convert('RGB')

result = pipeline.run(image)

print(f"Story:     {result['narrative_text'][:100]}...")
print(f"Emotion:   {result['emotion_label']}")
print(f"Embedding: {result['embedding_source']} (shape {result['emotion_embedding'].shape})")
print(f"Audio:     {result['audio'].shape[0]} samples @ {result['sample_rate']}Hz")
```

```python
# ===== Save & Play Audio =====
pipeline.save_audio(result['audio'], 'outputs/demo.wav')

# Play in notebook:
from IPython.display import Audio
Audio(result['audio'], rate=result['sample_rate'])
```
